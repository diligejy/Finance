```python
%matplotlib inline
import requests
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import itertools as it
from io import BytesIO
from tqdm import tqdm
```

```python
page = 1
url = 'https://finance.naver.com/sise/investorDealTrendDay.nhn?bizdate=20200622&sosok=&page={page}'.format(page=page)
```

```python
resp = requests.get(url)
html = BytesIO(resp.content)
```

```python
html
```

```python
df = pd.read_html(html)[0]
```

```python
cols = df.columns
```

```python
df = pd.DataFrame(columns = cols)
df
```

```python
for page in tqdm(range(1, 384)):
    url = 'https://finance.naver.com/sise/investorDealTrendDay.nhn?bizdate=20200622&sosok=&page={page}'.format(page=page)
    resp = requests.get(url)
    html = BytesIO(resp.content)
    df_single = pd.read_html(html)[0].dropna()
    df = pd.concat([df, df_single])
```

```python
df.reset_index(drop = True, inplace = True)
```

```python
df
```

```python
df.columns = df.columns.droplevel()
```

```python
df.to_csv('ant_vs_foreign.csv')
```

```python
sns.regplot('외국인', '기관계', df)
```

```python
comb_it = list(it.combinations(['개인', '외국인', '기관계'], 2))
```

```python
fig, axes = plt.subplots(ncols = 3)
fig.set_size_inches(15, 5)
for ax, comb in zip(axes, comb_it):
    sns.regplot(comb[0], comb[1], df, ax = ax)
fig.tight_layout()
```
