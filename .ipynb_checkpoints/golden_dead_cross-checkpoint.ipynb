```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import pandas_datareader as web
symbol = '005930.KS'
start = '2019-11-01'
end = '2020-01-01'
sec_prices = web.DataReader(symbol, 'yahoo', start, end)
```

```python
type(sec_prices.index)
```

```python
type(sec_prices)
```

```python
sec_prices.head()
```

```python
sec_close = pd.DataFrame(sec_prices['Close'])
```

```python
from copy import copy
my_sec_dt = copy(sec_close)
```

```python
# 5일 이동평균선 
move_avr_5 = sec_close.rolling(window=5).mean()
```

```python
my_sec_dt['MoveAverage 5Days'] = move_avr_5
```

```python
# 골든 크로스 추출
golden_con = (sec_close['Close'] > move_avr_5['Close'])
```

```python
sec_golden_dt = sec_close[golden_con]
```

```python
my_sec_dt['Golden Cross'] = sec_golden_dt
```

```python
# 데드크로스 추출
dead_con = (sec_close['Close'] < move_avr_5['Close'])
```

```python
sec_dead_dt = sec_close[dead_con]
```

```python
my_sec_dt['Dead Cross'] = sec_dead_dt
```

```python
my_sec_dt.plot(figsize = (16, 9))
```
